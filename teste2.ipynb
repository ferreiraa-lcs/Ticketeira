```python
class Evento:
    def __init__(self, nome, data, preco):
        self.nome = nome
        self.data = data
        self.preco = preco
    def resumo(self):
        return f"O evento {self.nome}, ocorrerá no dia {self.data}, por R${self.preco: .2f} a inteira!"
    def meia(self):
        return f"O preço da meia entrada é: R${self.preco*0.5: .2f}."

evento01 = Evento("Show do Ferrugem", "07/05/2026", 97.50)
print(evento01.resumo())
print(evento01.meia())
```
